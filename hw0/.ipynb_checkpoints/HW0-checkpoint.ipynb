Let's start by getting all our imports right:

```python
# Python 3.6
import numpy as np  # Used to perform efficient (and convenient) array and matrix operations.
from sklearn import datasets  # Used to load standard datasets.
from matplotlib import pyplot as plt  # Used to create plots.
import math  # Used for trigonometric functions, log, pow, etc.
print('Finished Imports')
```

Click on the cell immediately above, then press shift+enter or click Cell->Run Cells. The box beside 'In' will fill with a star while it is in the execution queue, and then fill with a number when it is finished. The numbers go up every time any cell is executed, and they can be used to infer execution order.

Next, we're going to load a dataset, and examine it:

```python
iris = datasets.load_iris()
print(dir(iris))  # The dir function allows us to see the attributes the object responds to.
```

```python
print(iris.data.shape)  # Outputs row, column format
```

This dataset is a numpy matrix where each row is one data instance, and each column represents one feature. Unfortunately, we can't plot in 4 dimensions, so let's slice the numpy matrix.

```python
new_data = iris.data[:, 0:2]
# Note that : slices across the entire dimension, and a:b slices [a,b)
# That is, including a, up to and not including b.
print(new_data.shape)
```

```python
xs = new_data[:, 0]
ys = new_data[:, 1]  # Technically, we didn't need to make new_data before slicing these columns,
# but it's good to know that you can slice arbitrary ranges in multiple dimensions at once.
plt.scatter(xs, ys)
plt.title('Made by: [ypeng6]')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

This is nice, but what are we looking at? The Iris Dataset consists of 150 measurements of 4 physical properties across 3 species of iris (a type of flower). If we change the color based on the classificiation, the groups are a little clearer:

```python
xs = new_data[:, 0]
ys = new_data[:, 1]
plt.scatter(xs, ys, c=iris.target)  # set the color based on the classification
plt.title('Made by: [ypeng6]')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

We can also use matplotlib to plot series data, or even multiple series. Note the use of the plot command, in place of scatter.  Also note that these commands can be combined in the same graphic. All commands before show() will appear in the same figure.

```python
xs = np.arange(100) * .5 - 10  # Creates a list of 100 values in intervals of .5, starting at -10.
# Numpy makes it very easy (and efficient) to do elementwise operations on large datasets.
y1s = [x*x for x in xs]  # If you are unfamiliar, this is called Python's "List Comprehension Syntax"
y2s = [math.sin(x)*x*x for x in xs]
plt.plot(xs, y1s, '.r', label='x^2')# The last argument is a formatting string
# https://matplotlib.org/users/pyplot_tutorial.html is a good starting point for more info.
plt.plot(xs, y2s, '-b', label='sin(x)*x^2')
plt.title('Made by: [ypeng6]')
plt.legend(loc='upper left')
plt.show()
```

# Concluding Notes about Notebooks
We do not plan to have you turn in any Jupyter Notebooks this semester, but that doesn't mean you shouldn't be using them! Even when the final hand-in is a regular Python file, Notebooks make a great development environment. Here are some nice ways to use Notebooks:
- Add cells to an execution queue by pressing (Shift+Enter) in one cell while another is executing!
- Take advantage of the Notebook format to separate concerns, like processing data, learning from data, debugging learned models, and plotting output. Nothing is worse than running a learning algorithm for a long time, only to lose the results because the plot wasn't formatted correctly! With Notebooks, individual pieces can be edited and rerun independently.
- Plot data all the time. As shown above, plots can frequently be generated in less than 5 lines of code, and they give a huge amount of information. In a Notebook, you can have exactly the plots you want when you want them.
- Always check the shape of your numpy matrices. It can save hours of development time if you realize early that you're using one matrix when you really wanted its transpose.
- Use numpy.matrix.dump and numpy.load to save and load data in local files. Use pickle.dump and pickle.load for generic datastructures.
- Write test code for individual functions. You only have to run the code when you want, so it won't slow down your main execution flow.
- Take advantage of Markdown formatting to leave long comments. This is a great place to document
    - Tutorials and resources you've found.
    - Intuitive explanations of what should be happening.
    - Worklogs and Large ToDo's.
    - Explanations for lay users if you publish your notebook. Speaking of...
- Publish as HTML, LaTeX, or PDF (some assembly may be required for the latter two) using File->Download As options. A great way to advertise to potential employers is to publish Machine Learning blog posts created from the HTML version of a Jupyter Notebook.
